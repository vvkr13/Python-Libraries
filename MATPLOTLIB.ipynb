**Q) The local ice cream shop keeps track of how much ice cream they sell versus the noon temperature on that day. Here are their figures for the last 12 days:**

```
import pandas as pd
```

```
IceCreamSales= {'temperature':[14.2,16.4,11.9,15.2,18.5,22.1,19.4,25.1,23.4,18.1],
                  'sales':[215,325,185,332,406,522,412,614,544,421]}
```

```
df=pd.DataFrame(IceCreamSales,columns=['temperature','sales'])
df
```

```
plt.scatter(df['temperature'],df['sales'],color='grey')
plt.xlabel('Temperature',fontsize=13)
plt.ylabel('Sales in $',fontsize=13)
plt.title('IceCreamSales',fontsize=15)
plt.grid()
plt.show()
```

```
def sine(x):
  return np.sin(x)
def cosine(x):
  return np.cos(x)
x=[i for i in range(1,20)]
y=[i for i in range(1,20)]
#plt.figure(1)
plt.subplot(2,2,1)
plt.plot(x,sine(x),'.-g')
plt.title('sine(x)')
#plt.figure(2)
plt.subplot(2,2,2)
plt.plot(x,cosine(x),'x--r')
plt.title('cosine(x)')
#plt.figure(3)
plt.subplot(2,2,3)
plt.plot(y,sine(y),'o-y')
plt.title('sine(y)')
#plt.figure(4)
plt.subplot(2,2,4)
plt.plot(y,cosine(y),'.--m')
plt.title('cosine(y)')
plt.tight_layout()
plt.show()
```

```
#bar plot can make comparision between different variables
Team=['IND','AUS','SA','SL','ENG','PAK']
Won=[25,27,19,20,11,9]
plt.barh(Team,Won,color='grey')
plt.title('WORLDCUP SERIES 2020',fontsize=15)
plt.xlabel('TEAM',fontsize=13)
plt.ylabel('MATCHES WON ',fontsize=13)
plt.show()
```

```
#pie chart
## displays the data in a circular graph.
##the peices of the graph are proportional to the fraction of the whole in each category
import matplotlib.pyplot as plt
expenses='Rent','Food','Phone','Fuel','Extras','savings'
prices=[2600,2000,300,1500,1000,3500]
explode=(0,0,0,0,0,0.05)
plt.pie(prices,labels=expenses,explode=explode,autopct='%1.2f%%')
plt.show()
```
